### **1.3 Типичная задача аналитика**

**Вопрос:** какие иконки мы можем убрать из верхнего ряда,
чтобы людям, привыкшим к программе, не стало больно,
а новопришедшим людям стало попроще?

#### **Перевод с языка бизнеса на язык данных**

**Вопрос:** какие иконки мы можем убрать из верхнего ряда,
чтобы людям, привыкшим к программе, не стало больно,
а новопришедшим людям стало попроще?

* На верхней панели есть ненужные иконки.
* Можно убрать ненужные иконки.
* Что такое ненужная иконка? Это иконка, на которую никто не нажимает.
* Нужно посчитать, сколько людей нажимает на каждую иконку.
* А если на каждую кто-то нажимает? Нужно выбрать отсечку:
убираем иконки, которые нужны менее чем 5% пользователей.
* Нужно посчитать, какая доля клиентов пользовалась каждой из
иконок.
* Что такое “пользовались”? Считаем, что человек воспользовался
иконкой, если хоть раз нажал на неё.

#### **Посчитать то, что требуется посчитать**
**Решение:** считаем для каждой иконки долю пользователей,
которые нажимали на неё хотя бы один раз. Выкидываем
иконки, которыми пользовалось меньше 5% людей.

#### **Выводы**

* Задачи для аналитики могут появляться и из вашей головы, и от
других людей.
* Если от результата задачи ничего не зависит, возможно, её не
стоит делать.
* Спланируйте заранее, как вы будете искать ответ на конкретный
вопрос.

### Что такое машинное обучение

* **Data Scientist**
Извлекает важную информацию из данных для решения реальных
задач и принятия решений.

* **Инженер машинного обучения**
Учит машину самостоятельно принимать решения и обосновывать
выводы.

Зачастую, этой информации слишком много для того, чтобы обработать её вручную и принять решение. Однако мы можем заставить машины **принимать решения за нас**. Главное - <u>обосновывать свои выводы</u>.\
Это и есть задача инженеров **Машинного обучения**: - научить машину самостоятельно анализировать информацию, принимать решения и обосновывать свои выводы.


### Где используется машинное обучение

Теперь посмотрим, где может применять машинное обучение.\
Впринципе, везде, где есть какие-либо данные. Важно понимать, что **Машинное обучение** как и любая другая технология призвана в первую очередь помогать людям и зарабатывать деньги.


#### Маркетинг
Возьмём, к примеру, **маркетинг**: Онлайн-маркетинг и персонализированная реклама.\
К примеру, совершая заказ в интернет-магазине, сервис порекомендует сопутствующий товар, который может нам также пригодиться. А затем, рекомендация ещё всплывёт в контекстной рекламе в соц.сетях.\
В данном примере алгоритмы **Машинного обучения** во-первых, порекомендовали товары, которые нам скорее всего понравятся, а во-вторых, соединили наш аккаунт в данном интернет-магазине с аккаунтом в соц.сетях для того, чтобы дополнительно рекламировать нам товар.

#### Креативность
Другое применение **Машинного обучения** - это т.н. *искусственная креативность*, когда приложение на основе *нейросетей* обрабатывает картинки, например, в стиле какого-нибудь известного художника. Это приложение использует *нейронные сети* для того, чтобы на основе существующей картинки генерировать новую.

#### Написание текстов
Кроме картинок *нейронные сети* могут генерировать предложения и даже целые тексты.\
В начале 2019 года была создана *нейронная сеть*, способная генерировать тексты любой длины. Эта нейронная сеть настолько хорошо генерировала новостные статьи и пресс-релизы, что компания, которая её изобрела решила не публиковать полную версию, а только урезанную.\
Но и урезанная версия довольно неплохо справляется со своей задачей. Убедимся в этом.

Давайте откроем ссылку на [Talk to Transformer](https://app.inferkit.com/demo) и выберем интересную тему из списка, например, *"Дебютный альбом Илона Маска"*.\
Видим как *нейронная сеть* работает и постепенно появляется текст про несуществующий дебютный альбом Илона Маска. Постепенно текст появляется потому, что *нейронная сеть* предсказывает только одно следующее слово.\

### Работа инженера машинного обучения
* Инженеры учат машину самостоятельно принимать решение.
* Для этого инженеры создают модель машинного обучения (*нейронные сети*).
* Модель должна быть качественной.

<u>*Модель машинного обучения*</u> - это такой сборник правил, по которым машина будет принимать некоторые решения, однако содержание данного соборника (сами правила) определяются машиной **самостоятельно**. Именно поэтому, это и называется **машинным обучением**.\
Машина самостоятельно учится принимать решения.\
Машина подбирает эти правила таким образом, чтобы примнимать правильные решения как можно чаще.\

Главная задача инженера **МЛ** - не просто создать модель, но и сделать так, чтобы она приносила пользу заказчику. При этом, когда мы говорим "пользу" - мы подразумеваем, чтобы она была **качественной** с т.з. заказчика. Т.е. модель должна решать его, т.е. заказчика, задачи. Для того чтобы сделать модель качественной, инженеру нужно организовать свою работу чётко и структурно.

### Как создать качественную модель?
Найти ответы на следующие вопросы:
* Какие задачи будет решать модель?
* Какие данные мы можем использовать?
* Какую именно модель мы будем обучать?
* Как мы проверим модель?

Для начала важно понять, а зачем вообще заказчик хочет эту модель? Какие задачи будет реашть модель.\
Далее - сбор необходимых данных. И провека, что они подходят для обучения и в т.ч. не нарушают никаких законов.\
И только после этого, мы можем приступить к созданию модели.\
А после создания модели, нам нужно проверить модель в деле, сделав несколько предсказаний. И часто проведя дополнительное тестирование.

Следование чётко определённому процессу несёт определённые преимущества:
- во-первых, можно представить работу над моделью ввиде чек-листа, ориентируясь на который, уже точно не забудем сделать что-то полезное для модели и для заказчика
- во-вторых, когда вы работаете над несколькими задачами одновременно, вы в любой момент времени знаете статус каждой задачи: что сейчас делается, и какие шаги будут предприняты.

3. **Как понять, что мы хорошо решили задачу?**\
Мы удаляем минимальное количество фотографий с собаками

Предположим, к нам обратилась компания, которая разрабатывает приложения "Собачий Instagram". Их целевая аудитория - ярые любители собак, которые выкладывают фотографии своих питомцев, а другие пользователи их оценивают. Но в приложении завелись "тролли", которые выкладывают фотографии своих кошек, чем очень нервируют пользователей-собачников. И те перестают пользоваться приложением.\

Если переложить цель на язык **машиного обучения**, то заказчик хочет получить модель, которая бы выполняла классификацию загруженных фотографий на два класса: кошки и собаки.

Показателем качества модели заказчик считает процент неправильно удалённых фотографий, т.е. таких фотографий, на которых в действительности изображены собаки, но наша модель посчитала что там кошки и удалила эту фотографию.\
Процент таких удалений должен быть как можно меньше, чтобы пользователи остались довольны.\
Это и будет основым требованием к нашей системе.

#### **Этап 2: Data Understanding**
На этом этапе важно понять:
1. **Какие данные нужны системе?**\
Фотографии, поскольку
приложение заказчика
работает с фотографиями.
2. **Что должно быть на фотографиях?**\
Кошки, чтобы знать, кого
удалять, и собаки, чтобы
знать, кого не удалять.
3. **Сколько нужно фотографий?**\
1000 фотографий кошек и
1000 фотографий собак.

Данные в **машинном обучении** играют самую важную роль. И без данных машинное обучение впринципе бесполезно.\
Какие данные нам нужны для работы нашей системы? Т.к. приложение заказчика работает с фотографиями, то логично предположить, что и наша модель должна работать с **фотографиями**.\
Что должно быть изображено на этих фотографиях.

Мы как инженеры **МЛ** должны научить модель принимать решения: изображена ли на фотографии кошка, или же собака.\
Для этого, мы должны выступить в роли учителя для машины и объяснить ей как ребёнку, что на этой фотографии изображена кошка, а на этой - собака.

Поэтому нам нужны в первую очередь фотографии кошек, чтобы модель знала, какие фотографии *удалять*.\
Кроме того, нам пригодятся фотографии собак, чтобы модель знала кого *не нужно* удалять.\
Заказчик сказал, что кроме кошек и собак на фотографиях больше никого нет, поэтому, мы этим и ограничимся.

Следующий немаловажный вопрос: а сколько нам нужно фотографий? **Машинное обучение** очень жадное до данных: чем больше - тем лучше.\
Поэтому, нам нужно как минимум 1000 фотогрфий кошек и 1000 фотографий собак. Но если есть больше, - то будет ещё лучше.\
Ещё один очень важный момент: количество фотографий кошек и фотографий собак должно быть примерно одинаковым. Тогда наша модель сможет их лучше отличать.

Вцелом, это все требования, которые мы предъявляем данным: 1000 фотографий кошек, 1000 фотографий собак и количество фотографий кошек примерно одинаково с количеством фотографий собак.
