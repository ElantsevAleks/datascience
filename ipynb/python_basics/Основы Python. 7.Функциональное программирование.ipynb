## **11.2** *Отличие функционального стиля от императивного*

**Основы функционального стиля программирования**
* Функция как объект
* Функции высшего порядка
* **`map`** и **`filter`**

Суть в том, что **функция - тоже является объектом**. И её тоже можно обрабатывать как простые данные: можно передавать как параметры, делать списки, сборки из этих функций, и оперировать с ними различными способами.
