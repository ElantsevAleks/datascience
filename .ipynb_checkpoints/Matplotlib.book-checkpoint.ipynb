## Subplots
***

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
x = np.linspace(0, 10, 10)
y1 = 4 * x
y2 = [i**2 for i in x]

fig, ax = plt.subplots(figsize = (8,6))
ax.set_title('Figures of dependancy')
ax.set_xlabel('x', labelpad=20, color='b') # labelpad - a space from line x to a name of this line
ax.set_ylabel('y1, y2')
ax.grid(which='major', lw=1.2)
ax.grid(which='minor', ls='--', color='gray', lw=.5)
ax.scatter(x, y1, c='red', label='y1 = x*4')
ax.plot(x, y2, label='y2 = x^2')
ax.legend()

ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)

# Here we added in line 8 additional parametrs labelpad and color.
```

## Location of .legeng()
***

```python
# .loc - location of a legend
locs = ['best', 'upper right', 'upper left', 'lower left', 'lower right', 'right',\
       'center left', 'center right', 'lower center', 'upper center', 'center']

plt.figure(figsize=(12,12))
for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            # Here we state a location of our diagram
            # 3 - quantity of rows
            # 4 - quantity of columns
            # i*4+j - a place where we want to locate our diagram
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j]) # a title of a diagram from the list
            plt.plot(x, y1, 'o-r', label='line 1') 
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
```

```python
# For more flexiable managing of a location of our legend, we can use an optional parametr - 
# bbox_to_anchor of .legend()
# For this parametr we can state a tuple for 2 or 4 elements
# bbox_to_anchor = (x, y, width, heght)
# bbox_to_anchor = (x, y)

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend(bbox_to_anchor=(1, .6))
```

```python
# Optional parametrs for .legend()
# fontsize - int or float or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
# frameon - True or False - a frame of a legend 
# framealpha - None or float - a transparency of a legend
# facecolor - None or str - painting of a legend
# edgecolor - None or str - painting of edges
# title
# title_fontsize

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.m', label='line 2')
plt.legend(fontsize=14, shadow=True, framealpha=1, facecolor='y', edgecolor='g', title='Legend')
```

```python
import matplotlib.gridspec as gridspec

x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]
fg = plt.figure(figsize=(7, 3), constrained_layout=True)
# Here we are creating an object of a class GridSpec
# We state a quantity of columns and rows, and our figure
gs = gridspec.GridSpec(ncols=2, nrows=1, figure=fg)
fig_ax_1 = fg.add_subplot(gs[0,0])
plt.plot(x, y1)
fig_ax_2 = fg.add_subplot(gs[0,1])
plt.plot(x, y2)

# An alternative way to create a GridSpec object is:
# gs = fg.add_grid_spec(1, 2)
```

### Style of an arrow in annotation
***
arrowprops takes dict with values from class FancyArrowPacth

- **Curve**:\
    name: - \
    params: None
- **CurveB**:\
    name: ->\
    params: head_length=.4, head_width=.2
- **BracketB**:\
    name: -\[\
    params: widthB:1.0, lengthB=.2, angleB=None
- **CurveFilledB**:\
    name: -|>\
    params: head_length=.4, head_width=.2
- **CurveA**:\
    name: <-\
    params: head_length=.4, head_width=.2
- **CurveAB**:\
    name: <->\
    params: head_length=.4, head_width=.2
- **CurveFilledA**:\
    name: <|-\
    params: head_length=.4, head_width=.2
- **CurveFilledAB**:\
    name: <|-|>\
    params: head_length=.4, head_width=.2
- **BracketA**:\
    name: ]-\
    params: widthA:1.0, lengthA=.2, angleA=None
- **BracketAB**:\
    name: ]-[\
    params: widthA:1.0, lengthA=.2, angleA=None, widthB:1.0, lengthB=.2, angleB=None
- **Fancy**:
    name: fancy\
    params: head_length=.4, head_width=.2, tail_width=.4
- **Simple**:\
    name: simple\
    params: head_length=.4, head_width=.2, tail_width=.2
- **Wedge**:\
    name: wedge\
    params: tail_width=.3, shrink_factor=.5

```python
plt.figure(figsize=(7, 5))
arrows = ['-', '->', '-[', '|-|', '-|>', '<-', '<->', '<|-', '<|-|>', 'fancy', 'simple', 'wedge']

bbox_properties = dict(boxstyle='round, pad=.2', ec='k', fc='w', ls='-', lw=1)

ofs_x = 0
ofs_y = 0

for i, ar in enumerate(arrows):
    if i == 6:
        ofs_x = .5
        
    plt.annotate(ar, xy=(.4+ofs_x, .92-ofs_y), xycoords='data', xytext=(.05+ofs_x, .9-ofs_y),\
                    textcoords='data', fontsize=17, bbox=bbox_properties, arrowprops=dict(arrowstyle=ar))
    if ofs_y == .75:
        ofs_y = 0
    else:
        ofs_y += .15
```

### Style of a connection of a line
***
connectionstyle

We can state a style of line, which connects dots (xy, xycoords). 
As a value a parametr connectionstyle takes an object of a class ConnectionStyle or string with parametrs, which listed and separated by comma.

- **Angle**:\
    name: angle\
    params: angleA=90, angleB=0, rad=0
- **Angle3**:\
    name: angle3\
    params: angleA=90, angleB=0
- **Arc**:\
    name: arc\
    params: angleA=0, angleB=0, armA=None, armB=None, rad=0.0
- **Arc3**:\
    name: arc3\
    params: rad=0.0
- **Bar**:\
    name: bar\
    params: armA=None, armB=None, fraction=.3, angle=None

```python
import math
fig, axes = plt.subplots(2, 3, figsize=(12, 7))
conn_style = ['angle, angleA=90, angleB=0, rad=0',
             'angle3, angleA=90, angleB=0',
             'arc, angleA=0, angleB=0, armA=0, armB=40, rad=.0',
             'arc3, rad=-1.0',
             'bar, armA=.0, armB=.0, fraction=.1, angle=70',
             'bar, fraction=-.5, angle=180']
for i in range(2):
    for j in range(3):
        axes[i, j].text(.1, .5, '\n'.join(conn_style[i*3+j].split(',')))
        axes[i, j].annotate('text', xy=(.2, .2), xycoords='data', xytext=(.7, .8), textcoords='data',
                            arrowprops=dict(arrowstyle='->', connectionstyle=conn_style[i*3+j]))
```

## General setting up
***

One way is to create an Axes element and modify a part of params.
It will be more convinience to use **inset_axes()** from **mpl_tollkits.axes_grid1.inset_locator**.
***

**Parametrs of a function inset_axes()**

- **parent_axes**:\
    type: axes\
    description: an object of master Axes
- **width**:\
    type: float or str\
    description: a width of an object, states in % from a master Axes or as a number
- **height**:\
    type: float or str\
    description: a width of an object, states in % from a master Axes or as a number
- **loc**:\
    type: int or string, optional (a default value is 1)
    description: a location of an object. Takes:\
                - 'upper right': 1
                - 'upper left': 2
                - 'lower left': 3
                - 'lower right': 4
                - 'right': 5
                - 'center left': 6
                - 'center right': 7
                - 'lower center': 8
                - 'upper center': 9
                - 'center': 10
- **bbox_to_anchor**:\
    type: tuple or matplotlib.transforms.BboxBase or optional\
    description: position and correlation of sides of an object. Format: left angle, bottom angle, width, length.
- **bbox_transform**:\
    type: matplotlib.transforms.Transform or optional\
    description: transformation of an object
- **borderpad**:\
    type: float or optional\
    description: backlash between bbox_to_anchor and an object

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
np.random.seed(123)
vals = np.random.randint(11, size=(7,7))
fig, ax = plt.subplots()
gr = ax.pcolor(vals)
axins = inset_axes(ax, width='7%', height='50%', loc='lower left',\
                  bbox_to_anchor=(1.05, .0, 1, 1), bbox_transform=ax.transAxes, borderpad=0)
plt.colorbar(gr, cax=axins)

# If you need, it's possible to modify a scale of a color line by class Tick.
```

For state your own scale you need to give a list to colorbar() function by a param `tick`.
An example:

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
np.random.seed(123)
vals = np.random.randint(11, size=(7,7))
fig, ax = plt.subplots()
gr = ax.pcolor(vals)
axins = inset_axes(ax, width='7%', height='50%', loc='lower left',\
                  bbox_to_anchor=(1.05, .0, 1, 1), bbox_transform=ax.transAxes, borderpad=0)
plt.colorbar(gr, cax=axins, ticks=[0,5,10], label='Value')
```

If you want to state a text instead of numbers you can use set_yticklabels() function.

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
np.random.seed(123)
vals = np.random.randint(11, size=(7,7))
fig, ax = plt.subplots()
gr = ax.pcolor(vals)
axins = inset_axes(ax, width='7%', height='50%', loc='lower left',\
                  bbox_to_anchor=(1.05, .0, 1, 1), bbox_transform=ax.transAxes, borderpad=0)
cbar = plt.colorbar(gr, cax=axins, ticks=[0,5,10], label='Value')
cbar.ax.set_yticklabels(['Low', 'Medium', 'High'])
```

## Vizualization of data

### Linear diagram

<pre>
plot([x], y, [fmt], *, data=None, **kwargs)
plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
</pre>

If we invoke plot() with one param like plot(y), we will get a diagram, where y-axis will have the values from this list and x-axis will have the indexes.

Let's see the arguments of plot():
- x, x2 ...:\
    an array of data for x-axis
- y, y2 ...:\
    an array of data for y-axis
- fmt: str\
    a form of a diagram: '[marker][line][color]'\
- **kwargs:\
    enlistment of a huge quantity for setting up of a diagram.
    
Some properties:
- **alpha**:\
    type: float\
    description: transparent
- **color or c**:\
    type: color\
    description: a color of a line
- **fillstyle**:\
    type: {'full', 'Left', 'right', 'bottom', 'top', 'none'}\
    description: a style of filling
- **Label**:\
    type: object\
    description: text label for a line
- **Linestyle** or **ls**:\
    type: {'-', '--', '-.', ':', '', (offset, ob-off-seq), ...}\
    description: a style of line
- **Linewidth** or **lw**:\
    type: float\
    description: a width of a line
- **marker**:\
    type: matplotlib.markers\
    description: a style of a marker
- **markeredgecolor** of **mec**:\
    type: color\
    description: a color of edge of a marker
- **markeredgewidth** or **mew**:\
    type: float\
    description: a width of edge of a marker
- **markerfacecolor** or **mfc**:\
    type: color\
    description: a color of filling of a marker
- **markersize** or **ms**:\
    type: float\
    description: a size of a marker

### Parametrs of fmt

- **marker**:\
    type: str\
    description: a type of a marker
        '.'	point marker
        ','	pixel marker
        'o'	circle marker
        'v'	triangle_down marker
        '^'	triangle_up marker
        '<'	triangle_left marker
        '>'	triangle_right marker
        '1'	tri_down marker
        '2'	tri_up marker
        '3'	tri_left marker
        '4'	tri_right marker
        's'	square marker
        'p'	pentagon marker
        '*'	star marker
        'h'	hexagon1 marker
        'H'	hexagon2 marker
        '+'	plus marker
        'x'	x marker
        'D'	diamond marker
        'd'	thin_diamond marker
        '|'	vline marker
        '_'	hline marker 
        
- **line**:\
    type: str\
    description: a style of a line
        '-'	solid line style
        '--'	dashed line style
        '-.'	dash-dot line style
        ':'	dotted line style
        
- **color**:\
    type: color\
    description: color of a line
        'b'	blue
        'g'	green
        'r'	red
        'c'	cyan
        'm'	magenta
        'y'	yellow
        'k'	black
        'w'	white

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
plt.figure(figsize=(12,7))
plt.plot(x, y1, 'o-r', alpha=.7, label='first', lw=5, mec='b', mew=2, ms=10)
plt.plot(x, y2, 'v-g', label='second', mec='r', lw=2, mew=2, ms=12)
plt.legend()
plt.grid(True)
```

### Fill between line and axis

For filling you can use fill_between()
<pre>
fill_between(x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)
</pre>

• **x**: array\
• **y1**: array or number, the first line\
• **y2**: array or number, the second line\
• **where**: bool array, optional. Coords for filling if they true\
• **step**: {'pre', 'post', 'mid'}\
• ****kwargs**: properties of Polygon

```python
x = np.arange(.0, 5, .01)
y = np.cos(x*np.pi)

plt.plot(x, y, c='r')
plt.fill_between(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.arange(.0, 5, .01)
y = np.cos(x*np.pi)

plt.plot(x, y, c='r')
plt.fill_between(x, y, where=(y>.7) | (y<-.7))
plt.xlabel('x')
plt.ylabel('y')
```

There are 500 points on the diagram above, so as you can see, it's a mess.\
For avoiding this you can stand a step for marker - param **markevery** takes:
• None - will show every point\
• N - every Nth point\
• (start, N) - every Nth point from start\
• slice (start, end, N) - every Nth point from start to end\
• [i, j, k, l] - will show just i, j, k, l points


```python
x = np.arange(.0, 5, .01)
y = np.cos(x*np.pi)

m_ev_case = [None, 10, (100, 30), slice(100, 400, 15), [0, 100, 200, 300], [10, 50, 100]]

fig, ax = plt.subplots(2, 3, figsize=(10, 7))
axs = [ax[i, j] for i in range(2) for j in range(3)]
for i, case in enumerate(m_ev_case):
    axs[i].set_title(str(case))
    axs[i].plot(x, y, 'o', ls='-', ms=7, markevery=case)
```

#### Cutting of a diagram

If you want to hide a part of your diagram you should use **masked_where()** from Numpy.

```python
x = np.arange(.0, 5, .01)
y = np.cos(x*np.pi)

y_masked = np.ma.masked_where(y < -.5, y)
plt.ylim(-1,1)
plt.plot(x, y_masked, lw=3)
```
